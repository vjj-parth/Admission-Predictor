```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
target=df[['Chance of Admit ']]
features=df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA','Research']]
```

```python
model=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=11)
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
df1=y_test.copy()
df1['Predicted Values']=y_predict
df1.columns=['Actual Values','Predicted Values']
df1
```

```python
sns.distplot(y_test,hist=False,label='y_test')
sns.distplot(y_predict,hist=False,label='y_predict')
plt.legend()
```

SCORE

```python
r2_score(y_test,y_predict) * 100
```

```python
model.score(x_test,y_test) * 100
```
